```python
#histogram of variants in VCF vs # of people

#isolate variant counts for one chromosome
from matplotlib import pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import pandas 
import numpy as np

col_names = ['Subject ID', 'ID', 'Gestalt ID', 'Vendor', 'Gender', 'File Type'] #represents names field for df
for i in range(1, 23):
    col_names.append('chr%s' % i) #appends chroms to names list
    
#adding on the extras
col_names.append('chrX') 
col_names.append('chrY')
col_names.append('chrM')

chromosome_counts = pandas.read_table( 
    'file_counts_20160615.txt', #the file we are reading in
    skiprows=1,#skip the first row
    names= col_names)#names of the columns

#chromosome_counts = 
chromosome_counts.set_index("File Type", drop=False, inplace = True) #sets index to first col
```

```python
chromosome_counts
```

```python
vcfs = chromosome_counts.loc['VCF']
vcfs
```

```python
#isolating chr1 data
vcfs['chr1']
```

```python
#step histogram
n, bins, patches = plt.hist(vcfs['chr1'], bins = 500, histtype = 'step')
plt.xlabel("Number of variants on chromosome 1")
plt.ylabel("# of people in each bin")
plt.axis([0, 3500000, 0, 80])
plt.title("Step Histogram of # of variants on chr1")
plt.show()
```

```python
#cumulative step histogram
n2, bin2s, patches2 = plt.hist(vcfs['chr1'], bins = 500, cumulative = True, histtype = 'step')
plt.xlabel("Number of variants on chromosome 1")
plt.ylabel("# of people in each bin")
plt.axis([0, 3500000, 0, 8000])
plt.title("Step Cumulative Histogram of # of variants on chr1")
plt.show()
```

```python
#scatter histogram
labels = range(500000, 3495000, 5000)
vcfs['bins'] = pandas.cut(vcfs.chr1, range(500000, 3500000, 5000), right=False, labels=labels) #xvals
y = vcfs.groupby('bins').agg({'chr1': [np.size]}) #series that contains y vals
vcfs_sorted = vcfs.sort_values('chr1')
plt.scatter(labels, y['chr1'])
plt.xlabel("Number of variants on chromosome 1")
plt.ylabel("# of people in each bin")
plt.axis([0, 3500000, 0, 80])
plt.title("Scatter Histogram of # of variants on chr1")
plt.show()


```
