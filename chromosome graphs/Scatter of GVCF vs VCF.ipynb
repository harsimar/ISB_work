```python
import pandas
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

col_names = ['Subject ID', 'ID', 'Gestalt ID', 'Vendor', 'Gender', 'File Type'] #represents names field for df
chroms = [] #just a list of chrom column labels (autosomes), to make selection easier
for i in range(1, 23):
    col_names.append('chr%s' % i) #appends chroms to names list
    chroms.append('chr%s' % i) 
    
#adding on the extras
col_names.append('chrX') 
col_names.append('chrY')
col_names.append('chrM')

chromosome_counts = pandas.read_table( 
    'file_counts_20160615.txt', #the file we are reading in
    skiprows=1,#skip the first row
    names= col_names)#names of the columns

#chromosome_counts = 
chromosome_counts.set_index("Vendor", drop=False, inplace = True) #sets index to first col
chromosome_counts.head()
```

```python
#isolating ILMN dataframe
ilmn = chromosome_counts.loc['ILMN']
```

```python
#pulling out gvcf vs vcf dataframes (of tech ILMN)
ilmn.set_index("File Type", drop=False, inplace = True)
gvcfs = ilmn.loc['GVCF']
vcfs = ilmn.loc['VCF']
```

```python
#creating the scatterplots

pp = PdfPages('GVCF_VCF variants.pdf')
for i in range(1, 23): #to produce the 22 graphs highlighting the points for 1 chromosome in a different color than the rest
    fig, ax = plt.subplots()
    ax.set_title('GVCF vs VCF variant counts for each person across all chr pairs: highlighting chr%d data' %i)
    ax.set_xlabel('GVCF variant counts')
    ax.set_ylabel('VCF variant counts (log scale)')
    ax.set_yscale('log')
    for chrNum in range(1, 23): #for one graph, iterating through GVCF dataframe
            plt.axis([0, 5 * 10**7, 0,  0.4 * 10**7])
            if chrNum == i:
                ax.scatter(x = gvcfs['chr%d' %chrNum], y = vcfs['chr%d' %chrNum], c = 'red', s = 15, marker = 'o', alpha = 1)
            else:
                ax.scatter(x = gvcfs['chr%d' %chrNum], y = vcfs['chr%d' %chrNum], c = 'blue', s = 5, marker = '.', alpha = 0.05)
                
    #plt.show()
    plt.savefig(pp, format='pdf') #saving one graph into pdf
pp.close()
```
