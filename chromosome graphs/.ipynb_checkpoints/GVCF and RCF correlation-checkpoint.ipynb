```python
import pandas
import numpy as np
import matplotlib.pyplot as plt

col_names = ['Subject ID', 'ID', 'Gestalt ID', 'Vendor', 'Gender', 'File Type'] #represents names field for df
chroms = [] #just a list of chrom column labels (autosomes) , to make selection easier
for i in range(1, 23):
    col_names.append('chr%s' % i) #appends chroms to names list
    chroms.append('chr%s' % i) 
    
#adding on the extras
col_names.append('chrX') 
col_names.append('chrY')
col_names.append('chrM')

chromosome_counts = pandas.read_table( 
    'file_counts_20160615.txt', #the file we are reading in
    skiprows=1,#skip the first row
    names= col_names)#names of the columns

#chromosome_counts = 
chromosome_counts.set_index("Vendor", drop=False, inplace = True) #sets index to first col
chromosome_counts.head()
```

```python
#isolating ILMN dataframe
ilmn = chromosome_counts.loc['ILMN']
```

```python
#pulling out gvcf vs vcf dataframes (of tech ILMN)
ilmn.set_index("File Type", drop=False, inplace = True)
gvcfs = ilmn.loc['GVCF']
rcfs = ilmn.loc['RCF']
```

```python
#trying to generalize calculation to every chromosome pair
gvcfs_chroms = gvcfs[chroms]
rcfs_chroms = rcfs[chroms]
matrix = np.empty((22, 22))
for i, col1 in enumerate(chroms):
    for j, col2 in enumerate(chroms):
        nd = np.corrcoef(gvcfs_chroms[col1], rcfs_chroms[col2])
        val = nd[0,1] #pulling the correct value from the matrix that is returned from above line
        matrix[i,j] = val #adding the value into my matrix
```

```python
#tweaking the matrix so differences are more noticable (big dif = big #)
matrix = matrix * 100
matrix = matrix.astype(int)
for i in range(22):
    for j in range(22):
        if matrix[i][j] >= 0: 
            matrix[i][j] = 100 - matrix[i][j] #distance from 100
        else: #if correlation negative
            matrix[i][j] = (-100) - matrix[i][j] #distance from -100
        
matrix
```

```python
fig, ax = plt.subplots()
heatmap = ax.pcolor(np.flipud(matrix), cmap = 'PuOr')

ax.grid(True)
x_location = np.arange(0.5, 23.0, 1)
x_labels = range(1, 23)
plt.xticks(x_location, x_labels)

y_location = x_location
y_labels = range(22, 0, -1)
plt.yticks(y_location, y_labels)


plt.axis([0, 22, 0, 22])
plt.title('Correlation of GVCF vs RCF (ILMN technology) for every chromosome pair')
plt.xlabel('Chromosome #')
plt.ylabel('Chromosome #')
#legend = plt.colorbar(heatmap)
#legend = fig.colorbar(heatmap, ticks=range(13, 34, 1))  

legend = fig.colorbar(heatmap, ticks= range(13, 34,1), orientation='horizontal')
legend.ax.set_xlabel('Correlation difference value: Value of 13 = .87 corrcoef, 33 = .67 corrcoef')
plt.show()
```
