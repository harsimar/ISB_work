```python
from matplotlib import pyplot as plt
import pandas
from matplotlib.backends.backend_pdf import PdfPages

col_names = ['Subject ID', 'ID', 'Gestalt ID', 'Vendor', 'Gender', 'File Type'] #represents names field for df
for i in range(1, 23):
    col_names.append('chr%s' % i) #appends chroms to names list
    
#adding on the extras
col_names.append('chrX') 
col_names.append('chrY')
col_names.append('chrM')

chromosome_counts = pandas.read_table( 
    'file_counts_20160615.txt', #the file we are reading in
    skiprows=1,#skip the first row
    names= col_names)#names of the columns

#chromosome_counts = 
chromosome_counts.set_index("File Type", drop=False, inplace = True) #sets index to first col
```

```python
chromosome_counts
```

```python
gvcfs = chromosome_counts.loc["GVCF"]
```

```python
gvcfs
```

```python
gvcfs_sorted = gvcfs.sort_values('chr1', ascending = False)
gvcfs_sorted
```

```python
gvcfs_sorted.set_index('Subject ID', drop = False, inplace = True)
gvcfs_sorted
```

```python
gvcfs_sorted['Index'] = range(1, len(gvcfs_sorted) + 1, 1)
gvcfs_sorted['Index']
```

```python
#graph of gvcfs_sorted with cg + ilmn vendors
pp = PdfPages('people_chr1_variant_counts.pdf')
ax = gvcfs_sorted.plot(kind = 'scatter', x = 'Index', y ='chr1', title = 'Variant Counts of Chr1 ~ GVCF file, all vendors')
ax.set_xlabel('Index of gvcfs_sorted dataframe')
ax.set_ylabel('Variant Counts in Chromosome 1')
plt.axis([0, 8000, 0, 6 * 10**7])
#plt.show()
plt.savefig(pp, format='pdf')
```

```python
gvcfs_sorted.tail(20)
```

```python
temp = gvcfs_sorted.set_index('Index', drop = False)
temp[temp['Vendor'] == 'CG'].index.tolist()
```

```python
ilmn_sorted = gvcfs_sorted[gvcfs_sorted['Vendor'] == 'ILMN']
cg_sorted = gvcfs_sorted[gvcfs_sorted['Vendor'] == 'CG']
cg_sorted
```

```python
ax = ilmn_sorted.plot(kind = 'scatter', x = 'Index', y ='chr1', title = 'Variant Counts of Chr1 ~ GVCF file, ILMN vendor')
ax.set_xlabel('Index of gvcfs_sorted dataframe')
ax.set_ylabel('Variant Counts in Chromosome 1')
plt.axis([0, 4850, 0, 6 * 10**7])
#plt.show()
plt.savefig(pp, format='pdf')
```

```python
ax = cg_sorted.plot(kind = 'scatter', x = 'Index', y ='chr1', title = 'Variant Counts of Chr1 ~ GVCF file, CG vendor')
ax.set_xlabel('Index of gvcfs_sorted dataframe')
ax.set_ylabel('Variant Counts in Chromosome 1')
plt.axis([4808, 7251, 0, 1500000])
#plt.show()
plt.savefig(pp, format='pdf')
```

```python
rcfs = chromosome_counts.loc["RCF"]
rcfs
```

```python
#rcf scatter plots
rcfs_sorted = rcfs.sort_values('chr1', ascending = False)
rcfs_sorted['Index'] = range(1, len(rcfs_sorted) +1, 1)
```

```python
#rcf plot of ILMN + CG
ax = rcfs_sorted.plot(kind = 'scatter', x = 'Index', y ='chr1', title = 'Variant Counts of Chr1 ~ RCF file, all vendors')
ax.set_xlabel('Index of rcf_sorted dataframe')
ax.set_ylabel('Variant Counts in Chromosome 1')
plt.axis([0, 8000, 0, 3 * 10**6])
plt.savefig(pp, format='pdf')
pp.close()
```
