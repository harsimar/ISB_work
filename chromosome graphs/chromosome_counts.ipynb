```python
from matplotlib import pyplot as plt
import pandas
import numpy as np

chroms = ['Subject ID'] #represents a list of one person's chromosome data
col_names = ['Subject ID', 'ID', 'Gestalt ID', 'Vendor', 'Gender', 'File', 'Type'] #represents names field for df
for i in range(1, 23):
    col_names.append('chr%s' % i) #appends chroms to names list
    chroms.append('chr%s' % i) #appends chroms to one person list
    
#adding on the extras
col_names.append('chrX') 
col_names.append('chrY')
col_names.append('chrM')
chroms.append('chrX')
chroms.append('chrY')
chroms.append('chrM')


chromosome_counts = pandas.read_table( 
    'file_counts_20160615.txt', #the file we are reading in
    skiprows=1,#skip the first row
    names= col_names)#names of the columns

#chromosome_counts = 
chromosome_counts.set_index("Subject ID", drop=False, inplace = True) #sets index to first col
```

```python
chromosome_counts.head()
```

```python
gvcf_index = chromosome_counts.set_index("File") #sets index to file col
```

```python
gvcf_index.head() #prints out the table with file as the index
```

```python
gvcfs = gvcf_index.loc["GVCF"] #returns rows labelled by "GVCF" (so gvcfs is a dataframe)
```

```python
gvcfs #this is what gvcfs looks like
```

```python
gvcf_chroms = gvcfs[chroms].set_index('Subject ID') #returns a new dataframe of just gvcf chromosome data that has the index set to 'Subject ID'
```

```python
gvcf_chroms.iloc[0] #isolates the first row with col labels
```

```python
current = gvcf_chroms.loc['102-01276-02'] #locating first row by Subject ID number (which is what the index has been set to)
```

```python
plt.xlabel('One person') # x axis label
plt.ylabel('Variant Counts of chromosomes') # y axis label 
plt.title('Bar graph of counts for one person') #title

y = current.dropna()
x = range(len(y))
width = 1/1.5

chrom_Labels = chroms[1:]
x_location = np.arange(0.5, 24.5, 1)
plt.bar(x, y , width, color = "blue")  
plt.xticks(x_location, chrom_Labels)

plt.show()
```
