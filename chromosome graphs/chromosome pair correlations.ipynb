```python
import pandas
import numpy as np
import plotly.plotly as py
import plotly.graph_objs as go
import matplotlib.pyplot as plt

col_names = ['Subject ID', 'ID', 'Gestalt ID', 'Vendor', 'Gender', 'File Type'] #represents names field for df
chroms = [] #list of just chroms, excluding m because that might cause errors later
for i in range(1, 23):
    col_names.append('chr%s' % i) #appends chroms to names list
    chroms.append('chr%s' % i)
    
#adding on the extras
col_names.append('chrX') 
col_names.append('chrY')
col_names.append('chrM')
chroms.append('chrX') 
chroms.append('chrY')

chromosome_counts = pandas.read_table( 
    'file_counts_20160615.txt', #the file we are reading in
    skiprows=1,#skip the first row
    names= col_names)#names of the columns

#chromosome_counts = 
chromosome_counts.set_index("Vendor", drop=False, inplace = True) #sets index to first col
chromosome_counts.head()
```

```python
#isolating ilmn technology df
ilmn = chromosome_counts.loc['ILMN']
```

```python
#making sure all of the entries are of the GVCF file type
ilmn_gvcf = ilmn[ilmn['File Type'] == 'GVCF']
chroms_only_df = ilmn_gvcf[chroms]
```

```python
#finding the correlation between each chromosome pair, returns df
corr_df= chroms_only_df.corr('pearson')
```

```python
#converting df of correlation coefs to a numpy array and optimizing it to highlight differences
corr_matrix = corr_df.as_matrix()
corr_matrix = corr_matrix * 100
corr_matrix = corr_matrix.astype(int)
for i in range(24):
    for j in range(24):
        corr_matrix[i][j] = 100 - corr_matrix[i][j]
corr_matrix
```
