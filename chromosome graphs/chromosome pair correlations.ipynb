```python
import pandas
import numpy as np
import matplotlib.pyplot as plt

col_names = ['Subject ID', 'ID', 'Gestalt ID', 'Vendor', 'Gender', 'File Type'] #represents names field for df
chroms = [] #list of just autosomes
for i in range(1, 23):
    col_names.append('chr%s' % i) #appends chroms to names list
    chroms.append('chr%s' % i)
    
#adding on the extras
col_names.append('chrX') 
col_names.append('chrY')
col_names.append('chrM')

chromosome_counts = pandas.read_table( 
    'file_counts_20160615.txt', #the file we are reading in
    skiprows=1,#skip the first row
    names= col_names)#names of the columns

#chromosome_counts = 
chromosome_counts.set_index("Vendor", drop=False, inplace = True) #sets index to first col
chromosome_counts.head()
```

```python
#isolating ilmn technology df
ilmn = chromosome_counts.loc['ILMN']
```

```python
#making sure all of the entries are of the GVCF file type
ilmn_gvcf = ilmn[ilmn['File Type'] == 'GVCF']
chroms_only_df = ilmn_gvcf[chroms]
```

```python
#finding the correlation between each chromosome pair, returns df
corr_df= chroms_only_df.corr('pearson')
```

```python
#converting df of correlation coefs to a numpy array and optimizing it to highlight differences
corr_matrix = corr_df.as_matrix()
corr_matrix = corr_matrix * 100
corr_matrix = corr_matrix.astype(int)
for i in range(22):
    for j in range(22):
        if corr_matrix[i][j] >= 0:
            corr_matrix[i][j] = 100 - corr_matrix[i][j]
        else:
            corr_matrix[i][j] = (-100) - corr_matrix[i][j]
corr_matrix
```

```python
fig, ax = plt.subplots()
heatmap = ax.pcolor(np.flipud(corr_matrix), cmap = 'Greens')

ax.grid(True)
x_location = np.arange(0.5, 23.0, 1)
x_labels = range(1, 23)
plt.xticks(x_location, x_labels)

y_location = x_location
y_labels = range(22, 0, -1)
plt.yticks(y_location, y_labels)


plt.axis([0, 22, 0, 22])
plt.title('Correlation coef between chromosome pairs with GVCF file type and ILMN technology')
plt.xlabel('Chromosome #')
plt.ylabel('Chromosome #')
    
#legend = plt.colorbar(heatmap) #another way of making a legend

#legend = fig.colorbar(heatmap, ticks=[0, 1, 2, 3, 4])
#legend.ax.set_ylabel('Correlation difference value: Caluculated by 100 - (100 * corrcoef)', rotation=270)
#legend.ax.get_yaxis().labelpad = 15     #vertical colorbar

legend = fig.colorbar(heatmap, ticks=[0, 1, 2, 3, 4], orientation='horizontal')
legend.ax.set_xlabel('Correlation difference value: Value of 0 = corrcoef 1, 4 = corrcoef .96')  # horizontal colorbar

plt.show()
```
